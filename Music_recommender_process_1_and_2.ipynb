# Process_part_1

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 2. find url and store it in a variable
url = "https://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
# 3. download html with a get request
response = requests.get(url)
response.status_code
```

```python
#response.content
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
#soup.select("#chart-position-1 > div.chart-content.col-xs-12.col-sm-8")
```

```python
soup.select("div.col-sm-8 p")
```

```python
titles = soup.select("div.col-sm-8 p")

title_list = []

for title in titles:
    title_text = title.find(class_="title").get_text()
    title_list.append(title_text)

print(title_list)
```

```python
len(title_list)
```

```python
artists = soup.select("div.col-sm-8 p")

artist_list = []

for artist in artists:
    artist_text = artist.find(class_="artist").get_text()
    artist_list.append(artist_text)

print(artist_list)
```

```python
len(artist_list)
```

```python
# each list becomes a column
music = pd.DataFrame({"title":title_list,
                       "artist":artist_list
                      })
```

```python
music.head(40)
```

```python
music.shape
```

```python
music.to_csv('music_top100.csv', index=False)
```

# Process_Part_ 2

```python
import pandas as pd
import random

top100_list = pd.read_csv('music_top100.csv')
top100_list
```

```python
import difflib

def recommend_song(user_song, top100_list):
    matching_songs = top100_list[top100_list['title'].apply(lambda x: difflib.SequenceMatcher(None, user_song.lower(), x.lower()).ratio() > 0.5)]
    
    if not matching_songs.empty:
        # Randomly select a different song from the matching songs
        random_song = top100_list[top100_list['title'] != user_song].sample(1)
        recommended_song = random_song.iloc[0]['title']
        recommended_artist = random_song.iloc[0]['artist']
        return f"Great! How about checking out '{recommended_song}' by {recommended_artist}?"
    else:
        return "Sorry, I don't have a recommendation at this time."

# User input for the song
user_input = input("Type a song title: ")

recommendation = recommend_song(user_input, top100_list)
print(recommendation)
```
